# Panorama do COVID-19 no Brasil

 A doença de coronavírus (COVID-19) é uma doença infecciosa causada pelo vírus SARS-CoV-2.

 A maioria das pessoas infectadas com o vírus apresentará doença respiratória leve a moderada e se recuperará sem a necessidade de tratamento especial. No entanto, alguns ficarão gravemente doentes e exigirão atenção médica. Idosos e pessoas com condições médicas subjacentes, como doenças cardiovasculares, diabetes, doenças respiratórias crônicas ou câncer, têm maior probabilidade de desenvolver doenças graves. Qualquer pessoa pode adoecer com COVID-19 e ficar gravemente doente ou morrer em qualquer idade. 

 O vírus pode se espalhar da boca ou nariz de uma pessoa infectada em pequenas partículas líquidas quando tosse, espirra, fala, canta ou respira. Essas partículas variam de gotículas respiratórias maiores a aerossóis menores. É importante praticar a etiqueta respiratória, por exemplo, tossir com o cotovelo flexionado, e ficar em casa e se auto-isolar até se recuperar, caso não se sinta bem.

 Segundo estudos realizado, cerca de 80,9% das pessoas que contraem o novo coronavírus são assintomáticos, o que causa maior preocupação sobre a propagação do vírus. Pois muitas pessoas podem transmitir a doença e nem saber que estão infectados. 
 
No entanto, 13,8% das pessoas terão infecções severas e precisarão de oxigênio. O restante das pessoas, que representam 4,7%, serão classificadas como infecções graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Visando a atual situação Chinesa com o coronavírus, tenho como objetivo elevar a consciência sobre o Covid-19 no Brasil.

### Dicionário de Variáveis
 * iso_code = Código ISO 3166-1 alfa-3 que define um padrão para representar os países com 3 letras.
 * continent = Continente que o país se localiza
 * location = Localização geográfica
 * date = Data de observação
 * total_cases = Total de casos confirmados de COVID-19.
 * new_cases = Novos casos confirmados.
 * new_cases_smoothed = Média de novos casos.
 * total_deaths = Número total de mortes causadas por COVID-19. 
 * new_deaths = Números de novas mortes causadas por COVID-19.
 * new_deaths_smoothed = Média de mortes novas mortes causadas por COVID-19 em 7 dias.
 * total_cases_per_million = Total de casos confirmados por 1.000.000 de pessoas.
 * new_cases_per_million = Novos casos confirmados por 1.000.000 de pessoas.
 * new_cases_smoothed_per_million = Média de 7 dias de novos casos confirmados por 1.000.000 de pessoas.
 * total_deaths_per_million = Total de morte causadas por COVID-19 por 1.000.000 de pessoas.
 * new_deaths_per_million = Novas mortes causadas por COVID-19 por 1.000.000 de pessoas.
 * reproduction_rate = Taxa de repodrução da infecção com COVID-19
 *  icu_patients = Números de paciente com COVID-19 em unidades de terapia intensivas no dia (UTI).
 * icu_patients_per_million = Números de paciente com COVID-19 por 1.00.00 de pessoas, em unidades de terapia intensivas no dia (UTI).
 * hosp_patients = Número de pacientes internados com COVID-19 em um dia.
 * hosp_patients_per_million = Número de pacientes por 1.000.000 internados com COVID-19 em um dia.
 * weekly_icu_admissions = Número entradas de pacientes com COVID-19 em UTIs em uma semana
 * weekly_icu_admissions_per_million = Npumero de entradas de pacientes com COVID-19 em UTIs por 1.000.000 de pessoas em 1 semana.
 * weekly_hosp_admissions = Número de entradas de pacientes em hospitais em 1 semana.
 * weekly_hosp_admissions_per_million = Número de entradas de pacientes em hospitais por 1.000.000 de pessaos em 1 semana.
 * total_tests = Total de testes realizados.
 * new_tests = Novos testes realizados.
 * total_tests_per_thousand = Total de testes realizados por 1.000.
 * new_tests_per_thousand = Novos testes realizados por 1.000.
 * new_tests_smoothed = Média de novos testes de COVID-19 por 1 semana. Em países que não divulgam os dados diariamente.
 * new_tests_smoothed_per_thousand = Média de novos teste de COVID-19 por 1 semana por 1.000 pessoas.Em países que não divulgam os dados diariamente.
 * positive_rate = Média de 1 semana de taxa positiva por teste realizado.
 * tests_per_case = Média de 1 semana de testes realizados por novo caso confirmado de COVID-19.
 * tests_units = Unidades usadas pelo local para relatar seus dados de teste. Um arui depaís não pode conter unidades misturadas. Todas métricas relacionadas aos dados de teste, usam um padrão especifico.
 * total_vaccinations = Número total de vacinas contra o COVID-19 aplicadas.
 * people_vaccinated = Número total de pessoas que receberam pelo menos uma dose de vacina.
 * people_fully_vaccinated = Número total de pessoas que receberam todas as doses prescritas pelo protocolo inicial de vacinação.
 * total_boosters = Número de doses de reforço contra COVID-19.
 * new_vaccinations = Novas doses de vacinas aplicadas contro COVID-19 por dia.
 * new_vaccinations_smoothed = Média de 1 semana de novas doses aplicadas da vacina contra COVID-19. Para países que não relatam dados de vacinação diariamente, assumimos que a vacinação mudou proporcionalmente durante os dias.
 * total_vaccinations_per_hundred = Total de vacinas aplicadas por 100 pessoas na população total.
 * people_vaccinated_per_hundred = Número total de pessoas que receberam pelo menos uma dose de vacina por 100 pessoas na população total.
 * people_fully_vaccinated_per_hundred = Número total de pessoas com ciclo de imunização compléto por 100 pessoas na população total.
 * total_boosters_per_hundred = Número total de doses de reforço aplicada á cada 100 pessoas na população.
 * new_vaccinations_smoothed_per_million = Média de 1 semana de novas doses de vacinação contra COVID-19 aplicadas por 1.000.000 de pessoas na população total.
 * new_people_vaccinated_smoothed = Média de 1 semana de pessoas que receberam a primeira dose da vacina.
 * new_people_vaccinated_smoothed_per_hundred = Média de 1 semana de pessoas que receberam a primeira dose da vacina por 100 pessoas a população total.
 * stringency_index = Indice de rigor do Governo: Medida com base em 9 indicadores, incluindo fechamento de escolas, fechamento de locais de trabalho e proibições de viagens. Com escala de 0 a 100, sendo 100 a mais rigorosa.
 * population = População total do país.
 * population_density = Número de pessoas dividido por área territorial em quilômetros quadrados.
 * median_age = Média de idade da população projetada pela ONU, para população em 2020.
 * aged_65_older = Parcela da população com 65 anos ou mais.
 * aged_70_older = Parcela da população com 70 anos ou mais.
 * gdp_per_capita = Produto interno bruto com relaçao de poder de compra em dólares. 
 * extreme_poverty = Parcela da população que vive em extrema pobreza, ano mais recente disponível desde 2010
 * cardiovasc_death_rate = Taxa de mortalidade por doenças cadiovasculares em 2017(número anual de mortes por 1000.000 pessoas).
 * diabetes_prevalence = Prevalência de diabetes (% da população de 20 a 79 anos) em 2017
 * female_smokers = Proporção de mulheres fumantes.
 * male_smokers = Proporção de homens fumantes.
 * handwashing_facilities = Parcela da população com instalações básicas para lavagem das mãos.
 * hospital_beds_per_thousand = Camas hospitalares por 1.000 pessoas. (ano mais recente 2010)
 * life_expectancy = Expectativa de vida de recem nascidos em 2019.
 * human_development_index = Um índice composto que mede o desempenho médio em três dimensões básicas do desenvolvimento humano – uma vida longa e saudável, conhecimento e um padrão de vida decente. (Valores de 2019)
 * excess_mortality = Diferença percentual entre o número relatado de mortes semanais ou mensais em 2020–2021 e o número projetado de mortes para o mesmo período com base nos anos anteriores.
 * excess_mortality_cumulative = Diferença percentual entre o número acumulado de mortes desde 1º de janeiro de 2020 e as mortes projetadas cumulativas para o mesmo período com base nos anos anteriores.
 * excess_mortality_cumulative_absolute = Diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores.
 * excess_mortality_cumulative_per_million = Diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores, por milhão de pessoas.
 
### Importando os dados do COVID-19


```
# importar as bibliotecas necessárias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('max_columns', None)
pd.set_option('display.max_rows', None)

# importar o dataset COVID-19 para um DataFrame

df = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
```

### Limpando nosso DataFrame deixando apenas os países.

```
# Eliminando entradas ausentes 

df1 = df.dropna(subset=['continent'], axis=0, inplace=False)
```

Com `df.dropna` eliminamos todas entradas que tem a coluna "continente" sem dado. Deixando apenas os países para conseguirmos estudar.

#### Verificando entradas.

```
# Ordenando datos mais recentes em ordem decrescente de mortes.

df1.loc[df1.date == '2023-01-20'].sort_values(by="total_deaths", ascending=False)
```

 Acima conseguimos visuazlizar que nosso DataFrame mudou copnforme queriamos, e podemos seguir com nosso estudo.

 ### Plotando Graficos dos 5 países com mais casos de mortes causadas por COVID-19.

```
# Plotando gráfico 

fig, ax = plt.subplots()

x = (df1.groupby(['location']).total_cases.sum().sort_values(ascending = False)[:5]).values
y = (df1.groupby(['location']).total_cases.sum().sort_values(ascending = False)[:5]).index

ax = sns.barplot(x = x, y = y, ax = ax, palette = 'tab10')
ax.grid(False)
ax.set_title('Países com maior numero de mortes de COVID-19')
ax.set_xlabel('Países')
ax.set_ylabel('Mortes')
```

Usando o grafico de barras. Conseguimos identificar, claramente a diferença entre os 5 países com mais mortes causadas por COVID-19. Com os Estados Unidos com uma quantidade de mortes, muito maior que o segundo país com mais mortes, que é a India. Em seguida vem Brasil em terceiro lugar.

#### Plotando Grafico de linha para visualizacao da evoluacao de casos no mundo.

```
# Criando variável df2 com apenas entradas World.

df2 = df.loc[df.location == "World"]

# Plotando gráfico de linhas.

fig = plt.gcf()
fig.set_size_inches(15, 5)
sns.set(style="whitegrid", palette = 'hot')

ax = sns.lineplot(
    data=df2, x="date", y="total_cases",
);

ax.grid(False)
ax.set_title("Casos de COVID-19 no mundo até {df.date.max()}")
ax.set_xlabel("Data")
ax.set_ylabel("Número de casos")

plt.show()
```

Usando o Gráfico de linhas, podemos obervar a curva de evolução de casos no mundo. Também uma possível correlação entre o periodo de festas e medidas de relaxamento do isolamento social, com aumento de casos. 

#### Plotando Grafico de linha para visualizacao da evoluacao de mortes causadas por COVID-19 no mundo.

```
# Plotando gráfico de linhas.

fig = plt.gcf()
fig.set_size_inches(15, 5)
sns.set(style="whitegrid", palette = 'hot')

ax = sns.lineplot(
    data=df2, x="date", y="total_deaths",
);

ax.grid(False)
ax.set_title("Mortes de COVID-19 no mundo até {df.date.max()}")
ax.set_xlabel("Data")
ax.set_ylabel("Número de casos")

plt.show()
```

Podemos idenficar que o crescimento de numeros de mortes causadas por COVID-19, está diminuindo. Por grande influência das doses de vacinas aplicadas.

### Plotando gráfico de dispersão entre a relacao entre o PIB do país e os casos de mortes.

Nesse gráfico podemos ver a relação entre a renda per capita e quantidades de mortes que um país teve. Conseguimos observar que mesmo alguns paises da europa que tem a renda mais alta, tiveram dificuldades com a pândemia. Os Estados Unidos mesmo sendo muito rico, foi o mais afetado em relação a quantidade de mortes.

```
# Plotando gráfico de dispersão.

df2 = df1.loc[df1.date >= '2023-01-18']
fig = plt.gcf()
fig.set_size_inches(15, 8)
ax = sns.scatterplot(data=df2, x="gdp_per_capita", y="total_deaths", hue="continent",  palette="tab10", s = 100)

ax.grid(False)
ax.set_title('PIB per capita do País x Número total de mortes em milhões')
ax.set_xlabel('PIB per capita')
ax.set_ylabel('Número total de mortes em milhões')
plt.show()
```

Vemos que independente da renda de uma país, aguns países ricos foram muito afetados com casos de mortes causadas por COVID-19. Porem, não podemos levar existem inumeros fatores que levaram a cada país terem seus desempenhos diferentes perante a pandemia.

### Vizualizando gráficos de linha da evolução de casos comos em escala linear e logarítmica. 

```
# Plotando gráficos de linha em escala linear e logarítmica

x = list(range(949))
x = df_brasil['date']
y = df_brasil['total_cases']

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 7))

# Gráfico 1
ax1.set_title('Casos no Brasil', Color='Grey', size=12, loc='left', pad=20)
ax1.set_xlabel('')
ax1.set_ylabel('Casos', size=11, color='dimgrey')
ax1.tick_params(axis='x', labelsize=11, colors='dimgrey')
ax1.tick_params(axis='y', labelsize=11, colors='dimgrey')
ax1.plot(x, y)

# Gráfico 2
ax2.set_title('Casos no Brasil (Escala logaritima)', color='Grey', size=12, loc='left', pad=20)
ax2.set_xlabel('')
ax2.set_ylabel('Casos', size=11, color='dimgrey')
ax2.tick_params(axis='x', labelsize=11, colors='dimgrey')
ax2.tick_params(axis='y', labelsize=11, colors='dimgrey')
ax2.set_yscale('log')
ax2.plot(x, y)

sns.despine()
plt.tight_layout()
```

### Vizualizando gráficos de linha da evolução de mortes comos em escala linear e logarítmica. 

```
# Plotando gráficos de linha em escala linear e logarítmica

x = list(range(949))
x = df_brasil['date']
y = df_brasil['total_deaths']

fig, (ax3, ax4) = plt.subplots(nrows=2, ncols=1, figsize=(10, 7))

# Gráfico 1
ax3.set_title('Mortes no Brasil', Color='Grey', size=12, loc='left', pad=20)
ax3.set_xlabel('')
ax3.set_ylabel('Mortes', size=11, color='dimgrey')
ax3.tick_params(axis='x', labelsize=11, colors='dimgrey')
ax3.tick_params(axis='y', labelsize=11, colors='dimgrey')
ax3.plot(x, y)

# Gráfico 2
ax4.set_title('Mortes no Brasil (Escala logaritima)', color='Grey', size=12, loc='left', pad=20)
ax4.set_xlabel('')
ax4.set_ylabel('Mortes', size=11, color='dimgrey')
ax4.tick_params(axis='x', labelsize=11, colors='dimgrey')
ax4.tick_params(axis='y', labelsize=11, colors='dimgrey')
ax4.set_yscale('log')
ax4.plot(x, y)

sns.despine()
plt.tight_layout()
```

Podemos observar nos gráficos de escala linear que apesar dos casos de COVID-19 estarem subindo praticamente em uma linha reta, as quantidades de mortes começaram a dar uma desacelerada.

Já no gráfico de escala logarítmica, vemos o crecimento muito esponencial de casos e mortes por COVID-19 no início da pandemia.

```
# Gráfico do número de mortes e vacinas por milhão de pessoas

fig, ax = plt.subplots(figsize=(14,5))

brasil_mortes = (df_brasil.loc[:, ['date', 'new_deaths_smoothed']].reset_index(drop=False))
vacinas = (df_brasil.loc[:, ['date', 'new_vaccinations_smoothed_per_million']].reset_index(drop=False))

sns.lineplot(x='date', y='new_deaths_smoothed', data=brasil_mortes, ci=None, linewidth=1, color='purple')
sns.lineplot(x='date', y='new_vaccinations_smoothed_per_million', data=vacinas, ci=None, linewidth=1, color='chartreuse')

ax.set_title('Número de mortes pelo COVID-19 x Vacinas por milhão de pessoas (Média móvel de 7 dias)', color='Grey', size=18, loc='left', pad=30)

ax.tick_params(axis='x', labelsize=14, colors='dimgrey')
ax.tick_params(axis='y', labelsize=14, colors='dimgrey')

ax.set_xlabel('')
ax.set_ylabel('')

sns.despine()
plt.tight_layout()
```

Nesse grafico conseguimos identificar uma clara relação entre as vacinas entre a redução das mortes causadas por COVID-19 e a aplicação da vacina. 
Após o inicio da vacinação as mortes começaram a diminuir e se manteve estavel.
